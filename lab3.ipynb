```python
import numpy as np
import pandas as pd

# Load the dataset
file_path = '"C:\Users\Utpal Raj\Downloads\DCT_withoutduplicate 5.csv"'
df = pd.read_csv(file_path)

# Assuming the LABEL column is the class label and we are dealing with two classes
# First, we isolate the features and the labels
X = df.iloc[:, :-1].values  # Features
y = df['LABEL'].values      # Labels

# Get unique classes
classes = np.unique(y)

# Check if there are at least two classes
if len(classes) < 2:
    raise ValueError("The dataset must contain at least two classes for this analysis.")

# Select two classes for the analysis
class_1 = classes[0]
class_2 = classes[1]

# Calculate the mean (centroid) for each class
centroid_1 = X[y == class_1].mean(axis=0)
centroid_2 = X[y == class_2].mean(axis=0)

# Calculate the spread (standard deviation) for each class
spread_1 = X[y == class_1].std(axis=0)
spread_2 = X[y == class_2].std(axis=0)

# Calculate the distance between the mean vectors of the two classes
distance_between_centroids = np.linalg.norm(centroid_1 - centroid_2)

# Display the results
print(f"Class 1 (Label: {class_1}) Centroid: {centroid_1}")
print(f"Class 1 (Label: {class_1}) Spread: {spread_1}")
print(f"\nClass 2 (Label: {class_2}) Centroid: {centroid_2}")
print(f"Class 2 (Label: {class_2}) Spread: {spread_2}")
print(f"\nDistance between the centroids of Class 1 and Class 2: {distance_between_centroids}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import minkowski

# Load the dataset
file_path = '/content/drive/MyDrive/DCT_withoutduplicate 5.csv'
df = pd.read_csv(file_path)

# Select two feature vectors (replace 'feature1' and 'feature2' with actual column names)
feature_vector_1 = df['feature1'].dropna().values
feature_vector_2 = df['feature2'].dropna().values

# Ensure that both vectors have the same length
min_length = min(len(feature_vector_1), len(feature_vector_2))
feature_vector_1 = feature_vector_1[:min_length]
feature_vector_2 = feature_vector_2[:min_length]

# Calculate Minkowski distances for r from 1 to 10
minkowski_distances = []
for r in range(1, 11):
    distance = minkowski(feature_vector_1, feature_vector_2, r)
    minkowski_distances.append(distance)

# Display the distances
print("Minkowski Distances for r from 1 to 10:", minkowski_distances)

# Plot the Minkowski distances
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), minkowski_distances, marker='o', color='b', linestyle='--')
plt.title('Minkowski Distance between Two Feature Vectors')
plt.xlabel('r (Minkowski Parameter)')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/MyDrive/DCT_withoutduplicate 5.csv'
df = pd.read_csv(file_path)

# Filter two classes, replace 'label' with your actual class label column
class1 = 'ClassA'  # Replace with actual class name
class2 = 'ClassB'  # Replace with actual class name
df_two_classes = df[df['label'].isin([class1, class2])]

# Separate the features (X) and labels (y)
X = df_two_classes[['feature1', 'feature2', 'feature3']]  # Replace with actual feature names
y = df_two_classes['label']  # Replace 'label' with actual label column

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize an empty list to store accuracies for different values of k
accuracies = []

# Loop over k values from 1 to 11
for k in range(1, 12):
    # Initialize the kNN classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)

    # Train the classifier on the training data
    knn.fit(X_train, y_train)

    # Predict on the test data
    y_pred = knn.predict(X_test)

    # Calculate the accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Store the accuracy
    accuracies.append(accuracy)

    # Print accuracy for each value of k
    print(f"Accuracy for k = {k}: {accuracy:.4f}")

# Plot the accuracy for k values from 1 to 11
plt.figure(figsize=(10, 6))
plt.plot(range(1, 12), accuracies, marker='o', linestyle='-', color='b')
plt.title('Accuracy vs. k in kNN Classifier')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.xticks(range(1, 12))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# Load the dataset
file_path = '/content/drive/MyDrive/DCT_withoutduplicate 5.csv'
df = pd.read_csv(file_path)

# Filter two classes, replace 'label' with your actual class label column
class1 = 'ClassA'  # Replace with actual class name
class2 = 'ClassB'  # Replace with actual class name
df_two_classes = df[df['label'].isin([class1, class2])]

# Separate the features (X) and labels (y)
X = df_two_classes[['feature1', 'feature2', 'feature3']]  # Replace with actual feature names
y = df_two_classes['label']  # Replace 'label' with actual label column

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the kNN classifier with k=3 (or any value you wish to experiment with)
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the training data
knn.fit(X_train, y_train)

# Predictions on training data
y_train_pred = knn.predict(X_train)

# Predictions on test data
y_test_pred = knn.predict(X_test)

# Confusion matrix for training data
conf_matrix_train = confusion_matrix(y_train, y_train_pred)
print("Confusion Matrix for Training Data:")
print(conf_matrix_train)

# Confusion matrix for test data
conf_matrix_test = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix for Test Data:")
print(conf_matrix_test)

# Classification report for training data (precision, recall, F1-score)
print("\nClassification Report for Training Data:")
print(classification_report(y_train, y_train_pred))

# Classification report for test data (precision, recall, F1-score)
print("\nClassification Report for Test Data:")
print(classification_report(y_test, y_test_pred))
```
